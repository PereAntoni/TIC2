# PROBLEMA DE LES REINES
  
El problema diu que s'han de posar n reines sobre un tauler de n x n amb la condició que cap no amenaçar a una altra.

Per resoldre el problema s'han d'usar una estructura adaptada al problema, un algoritme que generi possibles solucions i un algoritme que seleccioni les propostes vàlides.

En aquest cas s'usarà un algoritme backtracking. Es farà una poda a les propostes no vàlides, retallant l'ordre de l'algoritme

### OPCIONS VÀLIDES

Per si una llista conté opcions vàlides:

1. Cap reina ocuparà la mateixa columna. No hi pot haver dues reines a la matixa posició de la llista.
2. Cap reina ocuparà la mateixa fila. La llista no pot tenir cap nombre duplicat.
3. Les diagonals es calculen. cap raina podrà ocupar una posició on abs(RFil1-RFil2) != abs(RCol1-RCol2).
